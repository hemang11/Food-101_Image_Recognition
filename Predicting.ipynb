```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Put the model and image path here
model = tf.keras.models.load_model('Food101.hdf5')
img_path = 'ramen2.jpg' # img_path
```

```python
# Image Preprocessign and Conversion
from tensorflow.keras.preprocessing import image
img = image.load_img(img_path, target_size=(224, 224))
plt.imshow(img)
img_array = image.img_to_array(img)
img_batch = np.expand_dims(img_array, axis=0)
img_preprocessed = tf.keras.applications.efficientnet.preprocess_input(img_batch)
```

```python
# Image Prediction from a Trained MobileNet_V2
Categories[np.argmax(model.predict([img_preprocessed]))]
```
